```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv("titanic_data.csv")
```

```python
sns.countplot(x="Survived", data=titanic_data)
```

```python
sns.boxplot(x="Pclass", y="Age", data=titanic_data)
```

```python
sns.scatterplot(x="Fare", y="Age", hue="Survived", data=titanic_data)
```

```python
sns.jointplot(x="Survived", y="Fare", kind="scatter", data=titanic_data)
```

```python
sns.histplot(data=titanic_data, x="Fare")
plt.title("Distribution of Fare")
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Age", hue="Survived", data=titanic_data)
```

```python
sns.boxplot(x="Survived", y="Age", data=titanic_data)
```

```python
sns.boxplot(x="Pclass", y="Survived", data=titanic_data)
```
